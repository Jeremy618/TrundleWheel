## <font color='red'> Multivariate LSTM model </font>


```python
file_path = "C:\\Users\\jerem\\Documents\\TrundleWheel\\Data\\User1_Jeremy_Bezancon\\" #file path of the txt file
data = pd.read_csv(file_path+'all_datas_2'+'.csv', sep=';')

row_s, row_e = 0, 1000

dataset = np.vstack((data.iloc[row_s:row_e,1],data.iloc[row_s:row_e,2],data.iloc[row_s:row_e,3],
                     data.iloc[row_s:row_e,4],data.iloc[row_s:row_e,5],data.iloc[row_s:row_e,6],
                     data.iloc[row_s:row_e,7],data.iloc[row_s:row_e,8],data.iloc[row_s:row_e,9],
                     data.iloc[row_s:row_e,10],data.iloc[row_s:row_e,11])).T
print(dataset)
```

```python
def split_sequence(sequences,n_steps):
    X,y = list(),list()
    
    for i in range(len(sequences)):
        end_ix = i +n_steps
        if end_ix >len(sequences):
            break
        
        seq_x,seq_y = sequences[i:end_ix,:-1],sequences[end_ix-1,-1]
        X.append(seq_x)
        y.append(seq_y)
        
    return np.array(X),np.array(y)
```

```python
n_steps = 50

X,y = split_sequence(dataset,n_steps)
print(X.shape,y.shape)

for i in range(100):
    print(X[i],y[i])
```

#### <font color='chocolate'>define model</font>

```python
# n_steps = 50
n_features = X.shape[2]
x0 = Input(shape=(n_steps,n_features))
x1 = LSTM(50,activation='relu')(x0)
output = Dense(1)(x1)
```

```python
model = Model(inputs = x0,outputs = output)
model.compile(optimizer='adam',loss='mse')
```

```python
hist = model.fit(X,y,epochs=200,verbose=1)
```

```python
y_pred_tab = []
time = []
valid_distance = []

row_e = 10000
step = 200
N = int(row_e/step)


for row in range(0,row_e,step):

    selected_columns = data.iloc[row:row + n_steps, 1:11]
    x_input = selected_columns.to_numpy()
    x_input = x_input.reshape((1,n_steps,n_features))
    
    y_pred=model.predict(x_input,verbose=0)
    
    y_pred_tab.append(y_pred[0,0])
    time.append(data.iloc[row,0])
    valid_distance.append(data.iloc[row,11])
    


plt.plot(time, valid_distance, 'r-', label='Validation data', linewidth=0.5)
plt.plot(time, y_pred_tab, 'b-', label='Prediction',linewidth=0.5)
plt.title('Prediction and Validation Data')
plt.xlabel('time')
plt.ylabel('Distance')
plt.legend()
plt.show()
```
